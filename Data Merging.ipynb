Find the union of all column names

```python
column_names = []
for i in range(len(city_list)):
    col_names = pd.read_csv("../listings_inside_airbnb/"+city_list[i]+".csv").columns.values.tolist()
    column_names = column_names + col_names
    #print(city_list[i])
    #print(pd.read_csv("../listings_inside_airbnb/"+city_list[i]+".csv").columns.values.tolist())

union_colnames = set(column_names)   
union_colnames
```

```python
len(union_colnames)
```

Find the intersection of all column names

```python
intersection_colnames = []
for i in range(len(city_list)):
    col_names = pd.read_csv("../listings_inside_airbnb/"+city_list[i]+".csv").columns.values.tolist()
    if i == 0:
        intersection_colnames = col_names
    else:
        intersection_colnames = set(intersection_colnames).intersection(col_names)
    #print(city_list[i])
    #print(pd.read_csv("../listings_inside_airbnb/"+city_list[i]+".csv").columns.values.tolist())

intersection_colnames
```

```python
len(intersection_colnames)
```

Columns not available in every dataset:

```python
dif_cols = union_colnames - intersection_colnames
dif_cols
```

All datasets merged:

```python
for i in range(len(city_list)):
    read_data = pd.read_csv("../listings_inside_airbnb/"+city_list[i]+".csv")
    if i == 0:
        data = read_data
    else:
        data = pd.concat([data, read_data], axis=0)

data
```

Now we have airbnb_data dataframe with zipcode column with only 5-length strings, thus we should find number of rows with 5-length strings that cannot be converted to integers. We will find those rows with checking if each character in the string in each row is a number:

```python
len(airbnb_data['zipcode']) - sum(airbnb_data['zipcode'].str.isnumeric())
```

Only 3 rows have characters other than numbers, thus they will be deleted:

```python
airbnb_data = airbnb_data[airbnb_data['zipcode'].str.isnumeric()]
airbnb_data.shape
```
