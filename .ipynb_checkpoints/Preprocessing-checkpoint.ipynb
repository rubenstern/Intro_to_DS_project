```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import re
from sklearn.preprocessing import MultiLabelBinarizer, LabelBinarizer
import matplotlib.pyplot as plt
#sns.set(style="ticks", color_codes=True)
```

```python
listings_merged = pd.read_csv('data/merged_data.csv',index_col=0)
#listings = pd.read_csv('data/listings.csv')
```

```python
print('shape: ',listings_merged.shape)
print('columns: ',listings_merged.columns.values)
```

```python
listings_merged.head(2)
```

```python
to_delete = ['description','calendar_last_scraped','host_about','host_has_profile_pic','host_id','host_identity_verified',
            'host_location','host_name','host_neighbourhood','host_picture_url','host_thumbnail_url',
            'host_url','host_verifications','jurisdiction_names','last_review','last_scraped','latitude',
             'license','calendar_updated','country','country_code','first_review','host_since','listing_url',
             'longitude','market','medium_url','monthly_price','name','notes','picture_url','transit',
             'require_guest_phone_verification','require_guest_profile_picture','requires_license','scrape_id',
             'space','street','summary','thumbnail_url','weekly_price','xl_picture_url','has_availability',
            'host_acceptance_rate','neighborhood_overview','neighbourhood','security_deposit','square_feet',
            'cleaning_fee','calculated_host_listings_count','host_response_rate','is_location_exact',
             'neighbourhood_cleansed','experiences_offered','price_x','smart_location','city','available']
            
#we delete monthly price as it would it too easy to predict daily price based on monthly price
#same for weekly price

#Then, other columns are deleted either because their are almost missing for every rows,
##either because we found that they should not have influence on the price of the listings
for i in to_delete:
    del listings_merged[i]
```

```python
listings_merged.head(3)
```

```python
listings_merged = listings_merged.dropna()
```

```python
print('missing values ratio (in %): ')
print(listings_merged.isnull().sum()/listings_merged.shape[0])
```

```python
listings_merged.shape
```

```python
listings_merged.head(10)
```

```python
set(listings_merged['state'])
```

```python
listings_merged['state'].replace('Ca','CA',inplace=True)
listings_merged['state'].replace('ca','CA',inplace=True)
listings_merged['state'].replace('il','IL',inplace=True)
listings_merged['state'].replace('ny','NY',inplace=True)
listings_merged['state'].replace('secc Terrazas','CA',inplace=True)
```

```python
print('amenities')
amenities_list = listings_merged['amenities'].apply(lambda x: re.sub('[{}"]', '', x).split(','))
mlb_amenities = MultiLabelBinarizer()
amenities_binarized = mlb_amenities.fit_transform(amenities_list)
listings_merged['amenities: '+mlb_amenities.classes_] = pd.DataFrame(amenities_binarized)

to_binarize = ['bed_type','cancellation_policy','room_type','state','host_response_time']

for i in to_binarize:
    print(i)
    lb = LabelBinarizer()
    binarized = lb.fit_transform(listings_merged[i])
    listings_merged[i+': '+lb.classes_.astype(object)] = pd.DataFrame(binarized)

to_binarize_with_nan = ['property_type']
for i in to_binarize_with_nan:
    print(i)
    lb = LabelBinarizer()
    binarized = lb.fit_transform(listings_merged[i].apply(lambda x: str(x)))
    listings_merged[i+': '+lb.classes_.astype(object)] = pd.DataFrame(binarized)

print('instant_bookable')
lb = LabelBinarizer()
binarized = lb.fit_transform(listings_merged['instant_bookable'])
listings_merged['instant_bookable: t'] = pd.DataFrame(binarized)

print('host_is_superhost')
lb = LabelBinarizer()
binarized = lb.fit_transform(listings_merged['host_is_superhost'])
listings_merged['host_is_superhost: t'] = pd.DataFrame(binarized)
```

```python
listings_merged['price'] = listings_merged['price_y'].apply(lambda x: float(re.sub('[,]','',x[1:])))
del listings_merged['price_y']
listings_merged['extra_people'] = listings_merged['extra_people'].apply(lambda x: float(re.sub('[,]','',x[1:])))
```

```python
#We are deleting columns that have been binarized or that we won't use
to_delete = ['amenities','bed_type','cancellation_policy','room_type','state','host_response_time',
           'property_type','instant_bookable','host_is_superhost']

for i in to_delete:
    del listings_merged[i]
```
